```python
import networkx as nx
G = nx.Graph()
G.add_node('u')
G.add_node('v')
G.add_node('w')
G.add_edge('u','v')
G.add_edge('u','w')
G.add_edge('v','u')
G.add_edge('w','v')
nx.draw(G)
```

```python
G.add_node('p')
G.add_edge('v','p')
nx.draw(G)
```
